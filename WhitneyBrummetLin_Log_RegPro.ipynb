# Linear Regression

### Dataset without noise

```python
##Data (without noise)
data = [((1, -2), 1), ((1, -1), 3), ((1, 0), 5), ((1, 1), 7), ((1, 2), 9), ((1, 3), 11), ((1, 4), 13), ((1, 5), 15), ((1, 6), 17), ((1, 7), 19), ((1, 8), 21), ((1, 9), 23), ((1, 10), 25)]

F = lambda W, X: sum([W*X for W,X in zip(W,X)])

##Separate the X0 and X1 values from the dataset and print the tuples
X = [d for d, _ in data] 
print("(X0, X1): ", X)

##Separate the Y values from the dataset and print the values
Y = [y for _, y in data] 
print("Y: ", Y)

print("In: Y = W1X1 + W0X0")
```

Gradiant Descent

```python
##learning rate
a = 0.01

##beginning Weights
model = [0, 0]

iter = 0

#learning loop with appropriate amount of iterations
for _ in range(351):
  ##Y values given current weight values
  yPredicted = [F(model, x) for x in X]

  ##Empirical Loss measured by squared loss function 
  ## r is the value Y is supposed to be. y is the value calulated using the current learned wieghts
  Error = sum([(r-y)**2.0 for r, y in zip(Y, yPredicted)]) 
  
  ##derivative of the loss function to help calculate new weights
  deltaX0 = 1/len(Y)*sum([(-2*(r-y)*x[0]) for r, y, x, in zip(Y, yPredicted, X)])
  deltaX1 = 1/len(Y)*sum([(-2*(r-y)*x[1]) for r, y, x, in zip(Y, yPredicted, X)])
  
  ##gradient descent model calculation
  model = [model[0] - a * deltaX0, model[1] - a * deltaX1]
  
  #print information every 50 iterations
  if iter%50==0:
    print("Iteration: ", iter)
    print("Predicted Y Values: ", yPredicted)
    print("Empirical Loss by squared loss function: ", Error)
    print("derivative of the loss function for X0: ", deltaX0)
    print("derivative of the loss function for X1: ", deltaX1)
    print("New Weights: ", model)

    dataX1 = [x[1] for x, y in data]
    plt.plot(dataX1, yPredicted, "g")
    plt.plot(dataX1, Y, "g+")
    plt.show()
  iter += 1
```
